```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df=pd.read_csv("students_scores.csv")
```

```python
df.head()
```

```python
eng=df["english"]
eng
```

```python
eng.describe()
```

```python
plt.hist(eng)
```

```python
sns.histplot(eng,kde=True)
```

```python
#母平均の点推定
mu=np.mean(eng)
mu
```

```python
#母分散の点推定（不偏分散）
sigma2=np.var(eng,ddof=1)
sigma2
```

```python
#標準誤差(SE)
sigma=np.std(eng,ddof=1)
se=sigma/np.sqrt(len(eng))
se
```

```python
jiyuu=len(eng)-1
interval=stats.t.interval(0.95,jiyuu,loc=mu,scale=se)
interval
```
