```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("./sample_data/california_housing_train.csv")
```

```python
data.head()
```

```python
data_p=data["population"]
data_p
```

```python
data_p=np.array(data_p)
```

```python
#合計値
np.sum(data_p)
```

```python
#平均値
np.mean(data_p)
```

```python
#中央値
np.median(data_p)
```

```python
#標本分散
np.var(data_p,ddof=0)
```

```python
#不偏分散
np.var(data_p,ddof=1)
```

```python
#標本標準偏差
np.std(data_p,ddof=0)
```

```python
#不偏標準偏差
np.std(data_p,ddof=1)
```
