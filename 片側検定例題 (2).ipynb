```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from scipy import stats

%matplotlib inline
```

## 片側検定１

ダイエット薬の効能を検定するために、被験者10人について服用後の体重変化を記録

H0：μ=0（ダイエット薬に効果はない。体重の変化は0。）<BR/>
H1：μ<0（ダイエット薬には、体重減少の効果がある）←片側対立仮説、本当に示したい事柄<BR/>
https://ai-trend.jp/basic-study/hypothesis-testing/sided-test/

```python
# 被験者10人について服用後の体重変化を記録したデータ
data = pd.Series([1.1, -1, -1.5, -8.5, 0.9, 1.2, 5.9, -5.4, -7.3, -2.3])
plt.hist(data)
mean = np.mean(data)
print('平均値', mean)
```

```python
# 標準誤差の計算
sigma = np.std(data, ddof = 1)
se = sigma/np.sqrt(len(data))
# t値の計算
t_value = (mean - 0)/se
print(f"t値:{t_value:.3f}")
```

```python
# p値の計算と検定
df = len(data) - 1
alfa = stats.t.cdf(np.abs(t_value), df = df) 
p = 1 - alfa
print(f"p値:{p:.3f}")
if p < 0.05:
    print("体重変化に有意差がある。ダイエット薬には効果がある。")
else: 
    print("体重変化に有意差があるとはいえない。ダイエット薬には効果があるとはいえない。")
```

## 片側検定２

あるメーカーの電球は、寿命が2,000時間であると宣伝している。このスペックが本当であるかを確かめるため、この電球を様々なお店で合計15個購入しテストを行ったところ、寿命の平均値は1,900時間、標準偏差は150時間だった。この電球の寿命が正規分布に従うとき、この結果からこの電球の宣伝文句は問題ないと考えられるか。

H0：μ=2000（この電球の寿命は2,000時間である）<BR/>
H1：μ<2000（この電球の寿命は2,000時間よりも短い）←片側対立仮説、本当に示したい事柄<BR/>
https://bellcurve.jp/statistics/course/9414.html

```python
#データ数
size = 15
# 標本平均
mean = 1900
# 標準誤差の計算
sigma = 150
se = sigma/np.sqrt(size)
# t値の計算
t_value = (mean - 2000)/se
print(f"t値:{t_value:.3f}")
```

```python
# p値の計算と検定
df = size - 1
alfa = stats.t.cdf(np.abs(t_value), df = df) 
p = 1 - alfa
print(f"p値:{p:.3f}")
if p < 0.05:
    print("製品寿命に有意差がある。この電球の寿命は2,000時間よりも短い。")
else: 
    print("製品寿命に有意差があるとはいえない。この電球の寿命は2,000時間よりも短いとはいえない。")
```

## 片側検定３

ある島には非常に珍しい鳥が生息している。研究員がその鳥の数（羽）を1年間に10回調査したところ、平均25、不偏分散9であった。この結果から、この島には21を超える数の鳥が生息していると言えるかどうか検定せよ。なお、有意水準は5%とする。

H0：μ=21（生息数は平均21である）<BR/>
H1：μ>21（生息数は平均21を超える）←片側対立仮説、本当に示したい事柄<BR/>
https://bellcurve.jp/statistics/course/10004.html #他にも問題があるのでみてみてください。

```python
#データ数
size = 10
# 標本平均
mean = 25
# 不偏分散
disp = 9
se = np.sqrt(disp/size)
# t値の計算
t_value = (mean - 21)/se
print(f"t値:{t_value:.3f}")
```

```python
# p値の計算と検定
df = size - 1
alfa = stats.t.cdf(np.abs(t_value), df = df) 
p = 1 - alfa
print(f"p値:{p:.3f}")
if p < 0.05:
    print("生息数に有意差がある。生息数は平均21を超える。")
else: 
    print("生息数に有意差があるとはいえない。生息数は平均21を超えるとはいえない。")
```
